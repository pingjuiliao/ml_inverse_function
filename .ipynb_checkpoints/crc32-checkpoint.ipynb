```python
import random
import tensorflow as tf
import zlib
print(tf.__version__)
```

```python
# data= "12345678".encode('utf-8')
# cksum = zlib.crc32(data)
# print(cksum)
M = 20000
N = 4
def data_gen(m = M, n = N) :
    data_int  =  [ random.getrandbits(N*8) for _ in range(m) ]
    data_bits =  [ [ 1 if b == '1' else 0 for b in bin(num)[2:].rjust(n*8, '0') ] for num in data_int ]
    data_bits =  [ b[::-1] for b in data_bits]
    return  data_bits, data_int

random.seed(0)
X, X_int = data_gen()
print(X_int[0])
print(X[0])
print(bin(X_int[0]) )
```

```python

Y_int =  [ zlib.crc32(byte_str) for byte_str in [ n.to_bytes(N, byteorder='little') for n in X_int ] ] 

Y     =  [ [ 1 if b == '1' else 0 for b in bin(n)[2:].rjust(N*8) ] for n in Y_int ]
print(Y_int[:2])
print(Y[:2])
for i in range(10) :
    assert( int( ''.join([ str(i) for i in Y[i]]), 2) == Y_int[i] )
```

```python
TRAIN_RATIO = .8

M_train = int(round(TRAIN_RATIO * M))
M_test  = M - M_train
# X_train, X_test = Y[:M_train], Y[M_train:]
Y_test_raw = X[M_train:]
X_train = tf.convert_to_tensor(Y[:M_train], dtype=float)
X_test  = tf.convert_to_tensor(Y[M_train:], dtype=float)
Y_train = tf.convert_to_tensor(X[:M_train], dtype=float)
Y_test  = tf.convert_to_tensor(X[M_train:], dtype=float)

print(X_train.shape)
```

```python
ACTIVATION = "relu"
OUT_ACTIVATION = "linear"

model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, input_shape=[32,]),
    tf.keras.layers.Dense(128, activation=ACTIVATION),
    tf.keras.layers.Dense(128, activation=ACTIVATION),
    tf.keras.layers.Dense(128, activation=ACTIVATION),
    tf.keras.layers.Dense(N * 8, activation=OUT_ACTIVATION)
])
model.compile(optimizer='adam', loss='mse')
model.summary()
model.fit(X_train, Y_train, epochs=5000)

print("Done\n")
print("=" * 100 + "\n")
print("Activation Function in Hidden Layer : " + ACTIVATION)
print("Activation Function in Output Layer : " + OUT_ACTIVATION)
print("\n" + "=" * 100)
```

```python
raw_predictions = model.predict(X_test)
print(raw_predictions[0])

predictions = [ [ 1 if p >= 0.5 else 0 for p in pred ] for pred in raw_predictions ]
print(predictions[0])
print(Y_test_raw[0])
correct = sum([ y_hat == y for y_hat, y in zip(predictions, Y_test_raw)])
accuracy = correct / float(M_test)
print(correct)
print("accuracy : {}".format(accuracy) )
```
