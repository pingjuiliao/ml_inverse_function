```python
import base64
import random
import tensorflow as tf
print(tf.__version__)
```

```python
## Xs'
M = 5000
N = 6 # 6 Bytes
def data_gen(m = M , n = N) :
    X_int   = [ random.getrandbits(n*8) for _ in range(m) ]
    X_bits  = [ [ 1 if num & (1 << pow2) else 0 for pow2 in range(n*8) ] for num in X_int ]
    X_bytes = [ num.to_bytes(n, 'little') for num in X_int ] 
    return X_bits, X_bytes
random.seed(0)
X, X_bytes = data_gen()
print(X[:2])
print(X_bytes[:2])
```

```python
## Ys'
Y_bytes = [ base64.b64encode(s) for s in X_bytes ]
Y_int   = [ int.from_bytes(s, 'little')   for s in Y_bytes ]
Y       = [ [1 if num & 1 << pow2 else 0 for pow2 in range(8*8) ] for num in Y_int ]


print(Y_bytes[:2])
print(Y[:2])
print(base64.b64decode(Y_bytes[0]))
```

```python
M_train, M_test  = int(M*0.8), int(M*0.2)
assert M_train + M_test == M
# we are trying to reverse it
X_train = tf.convert_to_tensor(Y[:M_train], dtype=float) 
X_test  = tf.convert_to_tensor(Y[M_train:], dtype=float)
Y_train = tf.convert_to_tensor(X[:M_train], dtype=float)
Y_test  = tf.convert_to_tensor(X[M_train:], dtype=float)
Y_test_raw = X[M_train:]
# Y_train, Y_test  = tf.convert_to_tensor(X[:M_train], dtype=float), tf.convert_to_tensor(X[M_train:], dtype=float)
print("X_train: ", X_train.shape)
print("Y_train: ", Y_train.shape)
print(X_train[0])
```

```python
ACTIVATION = 'relu'
OUT_ACTIVATION = 'linear'

model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, input_shape=[64,]),
    tf.keras.layers.Dense(128, activation=ACTIVATION),
    tf.keras.layers.Dense(128, activation=ACTIVATION),
    tf.keras.layers.Dense(128, activation=ACTIVATION),
    tf.keras.layers.Dense(48, activation=OUT_ACTIVATION)
])
model.compile(optimizer='adam', loss='mse')
model.summary()

model.fit(X_train, Y_train, epochs=5000)
# predictions = [ [ 1 if bit_predict >= 0.5 else 0 for bit_predict in predict ] for predict in model.predict(X_test) ]
print("=" * 150)
print("")
print("Activation in Hidden Layer : " + ACTIVATION)
print("Activation in Output Layer : " + OUT_ACTIVATION)
print("\n" + "=" * 150)
```

```python
raw_predictions = model.predict(X_test)
predictions     =  [ [ 1 if p >= .5 else 0 for p in pred ] for pred in raw_predictions ]
print(raw_predictions[0])
print(predictions[0])
print(Y_test_raw[0])
# correct = sum([ int(''.join([str(b) for b in x]),2) == int(''.join([str(b) for b in y]), 2) for x,y in zip(predictions, Y_test)])
correct  = sum([ y_hat == y for y_hat, y in zip(predictions, Y_test_raw)])
accuracy = correct / float(M_test)
print("Correct cases : {}/{}".format(correct, M_test) )
print("accuracy      : {}%".format(accuracy* 100.0 )  )
```
